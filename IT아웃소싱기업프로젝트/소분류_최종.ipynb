```python
categories = [
    '웹사이트 개발', '앱 개발', '디자인 및 퍼블리싱', '데이터 수집 및 자동화', 'SEO 및 마케팅',
    '자동화 및 스크립팅', '보안 및 취약점 관리', '게임 개발', '데이터베이스 관리', '기획 및 문서화',
    '관리 시스템', '블록체인 및 암호화폐', 'IoT 및 임베디드 시스템', '제어 및 센서 시스템',
    '전자상거래 및 CMS', 'AI 및 머신러닝', '데이터 분석 및 시각화', '유지보수', '서버 관리',
    '하드웨어 및 IT지원', '기타'
]
```

```python
def parse_subcategories(subcategories):
    if isinstance(subcategories, str):
        try:
            return eval(subcategories)
        except:
            return [subcategories]
    return subcategories
```

```python
df['소분류'] = df['소분류'].apply(parse_subcategories)
```

```python
one_hot = pd.DataFrame(0, index=df.index, columns=categories)
for idx, subcategories in df['소분류'].items():
    for subcategory in subcategories:
        if subcategory in categories:
            one_hot.at[idx, subcategory] = 1
```

```python
df = pd.concat([df, one_hot], axis=1)
```

```python
etc_count = df['기타'].sum()
total_count = len(df)
etc_ratio = etc_count / total_count
```

```python
print(f"'기타' 비율: {etc_ratio:.2%}")
```
