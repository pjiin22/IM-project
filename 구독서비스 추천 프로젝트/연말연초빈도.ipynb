```python
import dask.dataframe as dd
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import numpy as np
from collections import Counter
from wordcloud import WordCloud

import warnings
warnings.filterwarnings('ignore')
```

```python
df_last = pd.read_csv('연말업종빈도.csv')
df_first = pd.read_csv('연초업종빈도.csv')
df_normal = pd.read_csv('보통업종빈도.csv')
```

---------------------------------------------------

```python
def compare_two_datasets(df1, df2, col1_name, col2_name):
    comparison_df = pd.merge(
        df1, df2, on='단어', how='outer', suffixes=(f'_{col1_name}', f'_{col2_name}')
    ).fillna(0)
    
    comparison_df.rename(
        columns={f'빈도_{col1_name}': col1_name, f'빈도_{col2_name}': col2_name}, inplace=True
    )
    
    comparison_df['freq_diff'] = comparison_df[col1_name] - comparison_df[col2_name]
    return comparison_df
```

```python
#연말 vs 연초
comparison_last_first = compare_two_datasets(df_last, df_first, '연말빈도', '연초빈도')

#연말 vs 보통
comparison_last_normal = compare_two_datasets(df_last, df_normal, '연말빈도', '보통빈도')

#연초 vs 보통
comparison_first_normal = compare_two_datasets(df_first, df_normal, '연초빈도', '보통빈도')
```

```python
comparison_last_first.sort_values(by='freq_diff', ascending=False)
```
