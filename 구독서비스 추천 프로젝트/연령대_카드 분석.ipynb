```python
import dask.dataframe as dd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df = pd.read_csv('df.csv')
```

```python
age = pd.read_csv('age.csv')
```

```python
age
```

```python
data = pd.merge(df, age, on='고객ID', how='inner') 
```

```python
data
```

```python
data.info()
```

```python
data = data[['연령대','거래활동기간','거래 밀도 지수','월평균 거래금액','월평균 거래건수','업종다양성지수','신규유입고객','총거래금액','총거래건수']]
```

```python
pd.options.display.float_format = '{:.1f}'.format

agg = {
    '거래활동기간': 'mean',
    '거래 밀도 지수': 'mean',
    '월평균 거래금액': 'mean',
    '월평균 거래건수': 'mean',
    '업종다양성지수': 'mean',
    '신규유입고객': 'sum',
    '총거래금액': 'mean',
    '총거래건수': 'mean'
}


result = data.groupby('연령대').agg(agg).reset_index()
```

```python
correlation_matrix = result.corr()  
age_correlation = correlation_matrix['연령대']
```

```python
age_correlation
```

```python
plt.figure(figsize=(10, 8))  
sns.heatmap(
    correlation_matrix,
    annot=True,  
    fmt=".2f",   
    cmap="coolwarm",
    cbar=True    
)
plt.title("Correlation Matrix Heatmap", fontsize=16)
plt.show()
```
