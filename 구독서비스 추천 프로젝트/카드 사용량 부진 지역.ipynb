# 카드 사용 부진

```python
# 1. 사분위수 계산
q1 = data['월평균 거래금액'].quantile(0.25).round(4)
q3 = data['월평균 거래금액'].quantile(0.75).round(4)
iqr = q3 - q1  # IQR 계산
upper_bound = q3 + 1.5 * iqr 

# 2. 상, 중, 하 분류
def categorize(row):
    if row['월평균 거래금액'] > upper_bound:
        return '상'
    else:
        return '일반'

data['등급'] = data.apply(categorize, axis=1)
```

```python
data['등급'].value_counts()
```

```python
pd.set_option('display.float_format','{:2f}'.format)
```

```python
import matplotlib.pyplot as plt

upper_bound = q3 + 1.5 * iqr

plt.figure(figsize=(8, 6))
plt.boxplot(data['월평균 거래금액'], vert=True, patch_artist=True, meanline=True, showmeans=True)
plt.title('월평균 거래금액', fontsize=14)
plt.axhline(y=upper_bound, color='red', linestyle='--', label=f'Upper Bound: {upper_bound:.2f}')
plt.ylabel('월평균 거래금액', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 부진지역 내 주요 소비 업종

```python
a = df_under.groupby('최다가맹점업종명')['고객ID'].count().reset_index(name = '방문고객수')
a.sort_values('방문고객수',ascending=False)

```

### 부진지역 내 주요 소비 업종 소비 비율
부진 지역 내 모든 고객

```python
total_visits_poor = a['방문고객수'].sum()
a['소비비율'] = a['방문고객수'] / total_visits_poor
a.sort_values('소비비율', ascending=False).head(5)
```

### 비부진지역 내 주요 소비 업종

```python
# 비부진지역
b = uu.groupby('최다가맹점업종명')['고객ID'].count().reset_index(name = '방문고객수')
b.sort_values('방문고객수', ascending=False)
```

### 비부진지역 내 주요 소비 업종 소비 비율

```python
total_visits_non_poor = b['방문고객수'].sum()
b['소비비율'] = b['방문고객수'] / total_visits_non_poor
b.sort_values('소비비율', ascending=False).head(5)
```

### 저 비율차이가 통계적으로 유의한가?
- 대응표본

```python
# 택시회사 Nan값 삭제
c1 = comp.dropna()
c1
```

```python
pd.set_option("display.float_format","{:.2f}".format)
```

```python
from scipy import stats

under = c1['소비비율_x']
up = c1['소비비율_y']

diff = up - under

# 정규성 검정
print(stats.shapiro(diff))
```

```python
# 정규성 만족 X, Wilcoxon Signed-Rank Test
stat, p_value = stats.wilcoxon(diff)
print(f"Wilcoxon 검정 통계량: {stat}, p-value: {p_value}")
```
